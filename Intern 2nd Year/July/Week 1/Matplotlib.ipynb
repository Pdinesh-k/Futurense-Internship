Pie Chart - Distribution of Pet Types


```python
import matplotlib.pyplot as plt

pets = ['Dog', 'Cat', 'Rabbit']
counts = [4, 3, 3]

plt.pie(counts, labels=pets, autopct='%1.1f%%')
plt.title('Distribution of Pet Types')
plt.show()
```

 Bar Chart - Average Age of Each Pet Type

```python
import numpy as np

pet_types = ['Dog', 'Cat', 'Rabbit']
average_ages = [5, 4, 2]

plt.bar(pet_types, average_ages, color=['blue', 'orange', 'green'])
plt.xlabel('Pet Type')
plt.ylabel('Average Age')
plt.title('Average Age of Each Pet Type')
plt.show()
```

Scatter Plot - Weight vs Height of Each Pet

```python

weights = [20, 5, 2, 25, 6, 1.5, 15, 7, 2.5, 22]
heights = [60, 30, 20, 65, 35, 15, 55, 40, 25, 62]
colors = ['blue', 'orange', 'green', 'blue', 'orange', 'green', 'blue', 'orange', 'green', 'blue']

plt.scatter(weights, heights, c=colors)
plt.xlabel('Weight (kg)')
plt.ylabel('Height (cm)')
plt.title('Weight vs Height of Each Pet')
plt.show()
```

Line Chart - Age vs Weight for Each Pet Type

```python

ages = [5, 3, 2, 7, 4, 1, 2, 5, 3, 6]
weights = [20, 5, 2, 25, 6, 1.5, 15, 7, 2.5, 22]


plt.plot(ages[:4], weights[:4], label='Dogs', marker='o')
plt.plot(ages[4:7], weights[4:7], label='Cats', marker='o')
plt.plot(ages[7:], weights[7:], label='Rabbits', marker='o')
plt.xlabel('Age')
plt.ylabel('Weight (kg)')
plt.title('Age vs Weight for Each Pet Type')
plt.legend()
plt.show()
```
